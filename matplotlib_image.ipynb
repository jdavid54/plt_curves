```python
fichier = 'greyscale.png'
img = mpimg.imread(fichier)
plt.imshow(img)
print(img.shape)
plt.show()
```

```python
img
```

```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray

rgb2gray(img)
```

```python
from skimage import color
from skimage import io
img = color.rgb2gray(io.imread('karim.jpg'))
plt.imshow(img)
plt.show()

#Notes: The weights used in this conversion are calibrated for contemporary CRT phosphors: 
# Y = 0.2125 R + 0.7154 G + 0.0721 B
```

```python
#Alternatively, you can read image in grayscale by:

from skimage import io
img = io.imread('karim.jpg', as_gray=True)
plt.imshow(img)
plt.show()
```
